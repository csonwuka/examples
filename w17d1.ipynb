## PRINCIPAL COMPONENT ANALYSIS
## Meaning of PCA

Principal Component Analysis (PCA) is a dimensionality reduction technique that transforms high-dimensional data into a lower-dimensional representation while retaining most of the important information (variance).

PCA finds new features (called principal components) that summarize the original data with minimal information loss.
A principal component is a new feature created as a linear combination of the original features.

## Importance of PCA
PCA technique is particularly useful in processing data where multi–colinearity exists between the features/variables.

### 1. High Dimensionality
PCA can be used when the dimensions of the input features are high (e.g. a lot of variables). As the number of features (dimensions) of a dataset increases, data becomes sparse. This problem is known as the curse of dimensionality. Reducing the number of input features while keeping as much useful information as possible.

### 2. Noise Reduction
PCA can be also used for denoising and data compression. Many datasets contain irrelevant or noisy features. PCA removes dimensions that contribute little variance, effectively filtering noise.

### 3. Visualization
It is easy to visualize 2D data and 3D data, but data in 4D and above is difficult to visualize. PCA allows you to project high-dimensional data (e.g., 100 features) into 2D or 3D for visualization.

## DATA SCIENCE WORKFLOW

```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
data = load_wine()
df = pd.DataFrame(data.data, columns=data.feature_names)
df.info()
```

```python
X = df
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print(X_scaled.shape)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
X_pca.shape
```
